#### ScreenResolution

From this feature I will extract two:
1. Resolution --> map to 'type' of HD (Full, Ultra HD)
2. Mark if Touchscreen

```python
data['ScreenResolution'].value_counts()
```

```python
data['screen_resolution'] = data['ScreenResolution'].apply(lambda x: re.search('\d{3,5}x\d{3,5}', x).group(0))
data['ScreenResolution'] = data['ScreenResolution'].apply(lambda x: re.sub('\d{3,5}x\d{3,5}', '', x))
```

```python
data['touch_screen'] = data['ScreenResolution'].apply(lambda x: 1 if 'touchscreen' in x.lower() else 0)
data['ScreenResolution'] = data['ScreenResolution'].apply(lambda x: re.sub('Touchscreen', '', x))
```

```python
data['touch_screen'] = data['touch_screen'].astype('O')
```

```python
data['hd_category'] = data['ScreenResolution'].apply(lambda x: 'Full HD' if 'full hd' in x.lower() else 
                                                           ('Quad HD' if 'quad hd' in x.lower() else 
                                                            ('Ultra HD' if 'ultra hd' in x.lower() else np.nan)))

data.drop(['ScreenResolution'], 1, inplace = True)
```

```python
data.fillna('NAN').pivot_table(index = 'screen_resolution', columns = 'hd_category', values = 'Product', aggfunc = 'count').fillna('')
```

<img src="https://dl4ih61pxf6wa.cloudfront.net/wp-content/uploads/2017/01/16142119/Illus_resolutions.png" style="width: 600px;"/>

```python
def map_resolutions(pixels):
    
    width = int(re.search('\d{3,5}', pixels).group(0))
    
    if width <= 1600:
        return 'HD'
    elif width < 2560:
        return 'FullHD'
    elif width < 3840:
        return 'QuadHD'
    elif width >= 3840:
        return 'UHD'   
```

```python
data['hd_category'] = data['screen_resolution'].apply(lambda x: map_resolutions(x), 1)

data.drop(['screen_resolution'], 1, inplace = True)
```

#### Ram

```python
data['Ram'].value_counts()
```

```python
data['ram'] = data['Ram'].apply(lambda x: re.sub('\D', '', x)).astype('int')
data.drop('Ram', 1, inplace = True)
```

#### Memory

```python
data[['Mem1', 'Mem2']] = data['Memory'].str.split('+', expand = True, )
```

```python
data['Mem2_HDD'] = data['Mem2'].apply(lambda x: x if isinstance(x, str) and 'hdd' in x.lower() else 0)
data['Mem2_HDD'] = data['Mem2_HDD'].apply(lambda x: re.sub('[a-zA-Z]', '', str(x))).astype('int')

data['Mem2_SSD'] = data['Mem2'].apply(lambda x: x if isinstance(x, str) and 'ssd' in x.lower() else 0)
data['Mem2_SSD'] = data['Mem2_SSD'].apply(lambda x: re.sub('[a-zA-Z]', '', str(x))).astype('int')

data['Mem2_Hybrid'] = data['Mem2'].apply(lambda x: x if isinstance(x, str) and 'hybrid' in x.lower() else 0)
data['Mem2_Hybrid'] = data['Mem2_Hybrid'].apply(lambda x: re.sub('[a-zA-Z]', '', str(x))).astype('float').astype('int')

map_tb_to_gb = {1 : 1024, 2 : 2048}
data.replace({'Mem2_HDD' : map_tb_to_gb}, inplace = True)
data.replace({'Mem2_SSD' : map_tb_to_gb}, inplace = True)
data.replace({'Mem2_Hybrid' : map_tb_to_gb}, inplace = True)
```

```python
data['Mem1_HDD'] = data['Mem1'].apply(lambda x: x if isinstance(x, str) and 'hdd' in x.lower() else 0)
data['Mem1_HDD'] = data['Mem1_HDD'].apply(lambda x: re.sub('[a-zA-Z]', '', str(x))).astype('float').astype('int')

data['Mem1_SSD'] = data['Mem1'].apply(lambda x: x if isinstance(x, str) and 'ssd' in x.lower() else 0)
data['Mem1_SSD'] = data['Mem1_SSD'].apply(lambda x: re.sub('[a-zA-Z]', '', str(x))).astype('int')

data['Mem1_fs'] = data['Mem1'].apply(lambda x: x if isinstance(x, str) and 'flash' in x.lower() else 0)
data['Mem1_fs'] = data['Mem1_fs'].apply(lambda x: re.sub('[a-zA-Z]', '', str(x))).astype('int')

data['Mem1_Hybrid'] = data['Mem1'].apply(lambda x: x if isinstance(x, str) and 'hybrid' in x.lower() else 0)
data['Mem1_Hybrid'] = data['Mem1_Hybrid'].apply(lambda x: re.sub('[a-zA-Z]', '', str(x))).astype('float').astype('int')

map_tb_to_gb = {1 : 1024, 2 : 2048}
data.replace({'Mem1_HDD' : map_tb_to_gb}, inplace = True)
data.replace({'Mem1_SSD' : map_tb_to_gb}, inplace = True)
data.replace({'Mem1_Hybrid' : map_tb_to_gb}, inplace = True)
```

```python
data['memory_SSD'] = data['Mem1_SSD'] + data['Mem2_SSD']
data['memory_HDD'] = data['Mem1_HDD'] + data['Mem2_HDD']
data['memory_hybrid'] = data['Mem1_Hybrid'] + data['Mem2_Hybrid']
data['memory_flash_storage'] = data['Mem1_fs']

data.drop(['Mem1_SSD', 'Mem1_HDD', 'Mem1_Hybrid', 'Mem1_fs', 'Mem1', 
           'Mem2_SSD', 'Mem2_HDD', 'Mem2_Hybrid', 'Mem2', 'Memory'], 1, inplace = True)

data.head()
```
