### 2. Explore target variable (price)

```python
fig, ax = plt.subplots(figsize = (15, 3), ncols = 3)

sns.distplot(data['price'], kde = False, ax = ax[0])
sns.distplot(np.sqrt(data['price']), kde = False, ax = ax[1])
sns.distplot(np.log(data['price']), kde = False, ax = ax[2])

ax[0].set_title('Original price feature')
ax[1].set_title('Square root of price')
ax[2].set_title('Natural logarithm of price')

ax[1].set_xlabel('sqrt(price)')
ax[2].set_xlabel('log(price)')

plt.show()
```

Seems there are no outliers, but the data is very skewed to the right. Normally would decide to predict the log or sqrt of y in order to force the target distribution to be more similar to a Gaussian. This time tho, as I'm running multiple models (linear regression, knn, svm, trees) I will leave it as is and see which models take better care of the log-normal looking curve

### 4. Explore numeric features

#### Distributions

```python
g = sns.FacetGrid(data[numerics].melt(), col = 'variable', sharex = False, sharey = False, col_wrap = 4, height = 3, aspect = 4/3)
g.map(sns.distplot, 'value', kde = False)

plt.show()
```

```python
print('Original number of observations:', data.shape[0])

data = data[(data['x'] > 3) & (data['y'] < 15) & (data['z'] < 10) & (data['z'] > 2) &
            (data['y'] != 0) & (data['z'] != 0) &
            (data['table'] < 70) & (data['table'] > 50) &
            (data['depth'] > 55) & (data['depth'] < 68.5) &
            (data['carat'] <= 3)]

print('After applying filters on numeric features:', data.shape[0])
```

```python
fig = plt.figure(figsize = (8, 5))
ax = Axes3D(fig)

sc = ax.scatter(xs = data['x'], ys = data['y'], zs = data['z'], s = 5, c = data['carat'], marker = 'o', alpha = 1)
ax.set_xlabel('X - length')
ax.set_ylabel('Y - width')
ax.set_zlabel('Z - depth')

plt.legend(*sc.legend_elements(), bbox_to_anchor=(1.05, 1), loc=2)
plt.title('Carat is also - correctly - dependant of size')
plt.show()
```

Decisions:
1. Multiply x-y-z and create a 'volume' feature
2. Use their squares as predictors (xy plot shows some sort of 2nd degree polynomial relationship)
3. Take square of carat too

```python
data['x*y*z'] = data['x'] * data['y'] * data['z']

data['x2'] = np.square(data['x'])
data['y2'] = np.square(data['y'])
data['z2'] = np.square(data['z'])

data['carat2'] = np.square(data['carat'])
```

```python
fig, ax = plt.subplots(figsize = (10, 4))
sns.heatmap(data.corr(), ax = ax, cmap = 'coolwarm', center = 0, annot = True, fmt = '.2g')
plt.title('\n')
plt.show()
```
