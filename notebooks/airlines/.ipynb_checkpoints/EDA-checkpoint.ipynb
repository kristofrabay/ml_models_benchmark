```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('../../data/airlines/train.csv')
train.columns = train.columns.str.lower()
train.drop(['unnamed: 0', 'id'], 1, inplace = True)

train.head()
```

```python
test = pd.read_csv('../../data/airlines/test.csv')
test.columns = test.columns.str.lower()
test.drop(['unnamed: 0', 'id'], 1, inplace = True)

test.head()
```

```python
numerics = train.select_dtypes(include = 'number').columns.tolist()
categoricals = train.select_dtypes(include = ['object', 'category']).columns.tolist()

print('Numeric features:', ', '.join(numerics), '\n')
print('Categorical features:', ', '.join(categoricals))
```

### 1. Check missing (NaN, null, inf) values

```python
print(train.shape)
print(test.shape)
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
print(train.shape)
print(test.shape)

train = train[train['arrival delay in minutes'].isnull() == False]
test = test[test['arrival delay in minutes'].isnull() == False]

print(train.shape)
print(test.shape)
```

### 2. Explore target variable (price)

```python
train['satisfaction'] = train['satisfaction'].astype('category')
test['satisfaction'] = test['satisfaction'].astype('category')
```

```python
fig, ax = plt.subplots(figsize = (10, 3), ncols = 2)
sns.countplot(train['satisfaction'], ax = ax[0], order = ['neutral or dissatisfied', 'satisfied'])
sns.countplot(test['satisfaction'], ax = ax[1], order = ['neutral or dissatisfied', 'satisfied'])
ax[0].set_title('Train')
ax[1].set_title('Test')
plt.tight_layout()
plt.show()
```

### 4. Explore numeric features

#### Distributions

```python
g = sns.FacetGrid(train[numerics].melt(), col = 'variable', sharex = False, sharey = False, col_wrap = 4, height = 3, aspect = 4/3)
g.map(sns.distplot, 'value', kde = False)
plt.show()
```

```python
g = sns.FacetGrid(test[numerics].melt(), col = 'variable', sharex = False, sharey = False, col_wrap = 4, height = 3, aspect = 4/3)
g.map(sns.distplot, 'value', kde = False)
plt.show()
```

```python
train = train[train['age'] <= 80]
test = test[test['age'] <= 80]

train = train[train['flight distance'] <= 4000]
test = test[test['flight distance'] <= 4000]

train = train[train['departure delay in minutes'] <= 200]
test = test[test['departure delay in minutes'] <= 200]

train = train[train['arrival delay in minutes'] <= 250]
test = test[test['arrival delay in minutes'] <= 250]
```

```python
plt.figure()
for i in train['satisfaction'].unique():
    sns.distplot(train[train['satisfaction'] == i]['age'], hist = False, label = i, kde_kws = {'linewidth' : 2.5})
 
plt.figure()
for i in train['satisfaction'].unique():
    sns.distplot(train[train['satisfaction'] == i]['flight distance'], hist = False, label = i, kde_kws = {'linewidth' : 2.5})
    
plt.figure()
for i in train['satisfaction'].unique():
    sns.distplot(train[train['satisfaction'] == i]['departure delay in minutes'], hist = False, label = i, kde_kws = {'linewidth' : 2.5})
    plt.ylim(0, 0.025)
    
plt.figure()
for i in train['satisfaction'].unique():
    sns.distplot(train[train['satisfaction'] == i]['arrival delay in minutes'], hist = False, label = i, kde_kws = {'linewidth' : 2.5})
    plt.ylim(0, 0.025)
```

#### Encoding the binaries now. Multiple category features will be one hot encoded later for everything except CatBoost

```python
train.describe(include = 'object')
```

```python
mapper = {'gender' : {'Female' : 0, 'Male' : 1}, 
          'customer type' : {'disloyal Customer' : 0, 'Loyal Customer' : 1}, 
          'type of travel' : {'Personal Travel' : 0, 'Business travel' : 1}}

train.replace(mapper, inplace = True)
test.replace(mapper, inplace = True)
```

```python
train.rename(columns = {'gender' : 'is_male', 'customer type' : 'is_loyal_customer', 'type of travel' : 'is_business_travel'}, inplace = True)
test.rename(columns = {'gender' : 'is_male', 'customer type' : 'is_loyal_customer', 'type of travel' : 'is_business_travel'}, inplace = True)
```

#### Correlations

```python
fig, ax = plt.subplots(figsize = (13, 8))
sns.heatmap(train.corr(), ax = ax, cmap = 'coolwarm', center = 0, annot = True, fmt = '.2g', annot_kws={"size" : 8})
plt.title('Correlation heatmap\n')
plt.show()
```

```python
mapper = {'satisfaction' : {'neutral or dissatisfied' : 0, 'satisfied' : 1}}

train.replace(mapper, inplace = True)
test.replace(mapper, inplace = True)
```

```python
train.rename(columns = {'satisfaction' : 'satisfied'}, inplace = True)
test.rename(columns = {'satisfaction' : 'satisfied'}, inplace = True)
```

```python
train.to_csv('../../data/airlines/train_cleaned.csv', index = False)
test.to_csv('../../data/airlines/test_cleaned.csv', index = False)
```
